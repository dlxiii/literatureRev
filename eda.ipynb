```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 144)

# !conda install --yes xlrd
```

```python
filename = 'data/TransportTimeScale/Exported Items.csv'
dfs = pd.read_csv(filename)
dfs.head()
```

```python
df=dfs.copy()
df.drop(columns=["Key",'Access Date', 'Access Date',"ISBN","ISSN","Url",
                 "Date","Date Added","Date Modified","Pages","Num Pages","Issue",
                 "Number Of Volumes","Journal Abbreviation","Volume",
                 "Short Title","Series","Series Number","Automatic Tags",
                 "Series Text","Series Title","Publisher","Place","Language",
                 "Rights",	"Type",	"Archive","Call Number","Extra",
                 "Editor","Series Editor","Translator","Link Attachments",
                 "Contributor","Attorney Agent","Book Author","Library Catalog",
                 "Cast Member","Commenter","Composer","Cosponsor",
                 "Counsel","Interviewer","Producer","Recipient","Archive Location",
                 "Reviewed Author","Scriptwriter","Words By","Guest",
                 "Number","Edition","Running Time","Scale","Medium",
                 "Artwork Size","Filing Date","Application Number",
                 "Assignee","Issuing Authority","Country","Meeting Name",
                 "Conference Name","Court","References","Reporter",
                 "Legal Status","Priority Numbers","Programming Language",
                 "Version","System","Code","Code Number","Section",
                 "Session","Committee","History","Legislative Body"],
       inplace=True)
df.head(1)
```

```python
def addIndex(index_name):
    Mtags = df["Manual Tags"].values.tolist()
    len(Mtags)
    for i in df.index:
        if index_name in Mtags[i]:
            df.loc[i, index_name] = 1
        else:
            df.loc[i, index_name] = 0
```

```python
index_list = ["water age", "residence time", "average age", 
              "export time", "exposure time", "flushing time",
              "renewal time", "retention time", "transit time",
              "turnover time",
              "return coefficient","return-flow factor",
              "mean velocity", "vertical transport time"]
for index_name in index_list:
    addIndex(index_name)
```

```python
print (df.sum())
```

```python
df.to_csv('data.csv',index=False)
```

```python
index_sum = [df[i].sum() for i in index_list]
index_sum
```

```python
colors = sns.color_palette('pastel')[0:len(index_list)]
plt.pie(index_sum, labels = index_list, colors = colors, autopct='%.0f%%')
plt.savefig("plot_index.png",dpi=300)
plt.show()
```

```python
plt.clf()
plt.figure(figsize=(3,6))
sns.barplot(x=index_sum, y=index_list)
plt.savefig("plot_index_2.png",dpi=300)
```
